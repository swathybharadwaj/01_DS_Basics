# 2 - Identify variable types

## Integer Data Type

```python
data.dtypes[data.dtypes =='int64']
```

### NOTES
* Customer_id, vintage, age is okay for integer
* cutsomer_nw_category (ordinal) and branch_code (nominal) and churn (binary) (nominal) must be converted to categorical variable

```python
data.customer_nw_category.value_counts()
```

```python
data.branch_code.value_counts().head()
```

### Converting to categorical datatype

```python
data['customer_nw_category']=data['customer_nw_category'].astype('category')
data['branch_code']=data['branch_code'].astype('category')
data['churn']=data['churn'].astype('category')
```

## Float Variables

```python
data.dtypes[data.dtypes=='float64']
```

### NOTES
* City is a category (nominal) as there are various cities
* dependents must be a whole number. so would need to be converted to integer

```python
#data['dependents'].fillna(-1)
data['city']=data['city'].astype('category')
data['dependents']=data['dependents'].astype('Int64')
```

## Object Variables

```python
data.dtypes[data.dtypes=='object']
```

### NOTES

* Gender and occupation are categorical variables
* last_transaction must be  date time variable

```python
data['gender']=data['gender'].astype('category')
data['occupation']=data['occupation'].astype('category')
```

## Datetime Variables

```python
date=pd.DatetimeIndex(data['last_transaction'])
```

```python
# Extracting date features using date object created. Breaking down the last transaction will help us in understanding 
# when the last transaction was done from different perspectives
# Note - We dont need year as all data is from 2019 year
data['day_of_year']=date.dayofyear
data['week_of_year']=date.weekofyear
data['month_of_year']=date.month
data['date']=date.day
data['day_of_week']=date.dayofweek
```

```python
data.head()
```

```python
# Now that we have all features of date, we can drop last_transaction column

data.drop('last_transaction',inplace=True,axis=1)
```
